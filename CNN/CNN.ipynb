```python
import numpy as np
import pandas as pd 
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import csv
from sklearn import preprocessing
```

```python
train_images = pd.read_csv('/kaggle/input/digit-recognizer/train.csv').to_numpy()
test_images = pd.read_csv('/kaggle/input/digit-recognizer/test.csv').to_numpy()
```

```python
train_labels = train_images[:, 0]
train_images = train_images[:, 1:]
train_labels
```

```python
train_images = train_images.reshape(42000, 28, 28, 1)
test_images = test_images.reshape(28000, 28, 28, 1)
```

```python
train_images, test_images = train_images / 255.0, test_images / 255.0
```

```python
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
```

```python
model.summary()
```

```python
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              metrics=['accuracy'])
```

```python
history = model.fit(train_images, train_labels, epochs=100)
```

```python
predicted_labels = model.predict(test_images)
```

```python
predicted_labels
```

```python
file = open('submission.csv', 'w')
writer = csv.writer(file)

writer.writerow(["ImageId","Label"])
for i in range(len(predicted_labels)):
    writer.writerow([i+1, np.argmax(predicted_labels[i])])

file.close()
```
